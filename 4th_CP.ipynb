```python
#Correlation map
def plot_correlation_map( df ):
    corr = df.corr()
    s , ax = plt.subplots( figsize =( 20 , 20 ) )
    cmap = sns.diverging_palette( 220 , 10 , as_cmap = True )
    s = sns.heatmap( corr, cmap = cmap,square=True, cbar_kws={ 'shrink' : .9 }, ax=ax, annot = True, annot_kws = { 'fontsize' : 12 } )
plot_correlation_map(df)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from sklearn import metrics
```

```python
# SPLIT DATA
from sklearn.model_selection import train_test_split

x=df['sqft_living']
y=df['price'].values
x_train , x_test , y_train , y_test= train_test_split(x , y , test_size=0.20 , random_state=1)
x_train = np.array(x_train).reshape(-1,1)
y_train = np.array(y_train).reshape(-1,1)
x_test = np.array(x_test).reshape(-1,1)
y_test = np.array(y_test).reshape(-1,1)
```

```python
# Applying linear regression to our training set
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

model=LinearRegression()
model.fit(x_train,y_train)
predicted=model.predict(x_test)
print('Accurancy' , model.score(x_test , y_test) )
print("R squared =", metrics.r2_score(y_test,predicted))
print("MSE =", mean_squared_error(y_test,predicted))
```

```python
# we notice that our model predicted just 50% correctly that means that our model is performing badly and for the reason that we gave it just one feature
# as an input - which isnt enough - 
```

```python
# Apply multi-linear regression and compare it to the linear model

x=df[['bedrooms', 'bathrooms', 'sqft_living', 'sqft_lot',
       'floors', 'waterfront', 'view', 'condition', 'grade', 'sqft_above',
       'sqft_basement', 'yr_built', 'yr_renovated', 'zipcode', 'lat', 'long',
       'sqft_living15', 'sqft_lot15']]
y=df['price'].values

x_train , x_test , y_train , y_test= train_test_split(x , y , test_size=0.20 , random_state=1)

model=LinearRegression()
model.fit(x_train,y_train)
predicted=model.predict(x_test)
print('Accurancy' , model.score(x_test , y_test) )
print("R squared =", metrics.r2_score(y_test,predicted))
print("MSE =", mean_squared_error(y_test,predicted))
```

```python
# we notice that our accurancy increased when we used many features as an input instead of one 
# in other terms we gave our model enough features so it can discover the hidden pattern between the inputs and the output 
# we can say that 68.58% is quiet acceptable
# so as a conclusion the multi-linear regression gave us better prediction (a better model) than the linear regression (68%>50%)
```

```python
# Apply Polynomial regression and compare it to linear and multilinear regression. 
from sklearn.preprocessing import PolynomialFeatures

x=df[['bedrooms', 'bathrooms', 'sqft_living', 'sqft_lot',
       'floors', 'waterfront', 'view', 'condition', 'grade', 'sqft_above',
       'sqft_basement', 'yr_built', 'yr_renovated', 'zipcode', 'lat', 'long',
       'sqft_living15', 'sqft_lot15']]
y=df['price'].values

x_train , x_test , y_train , y_test= train_test_split(x , y , test_size=0.20 , random_state=1)

model=LinearRegression()
poly= PolynomialFeatures(degree=2)

x_train_fit=poly.fit_transform(x_train)
model.fit(x_train_fit,y_train)
x_test_fit=poly.fit_transform(x_test)
predicted=model.predict(x_test_fit)

print('Accurancy' , model.score(x_test_fit , y_test) )
print("R squared =", metrics.r2_score(y_test,predicted))
print("MSE =", mean_squared_error(y_test,predicted))
print("Intercept =" , model.intercept_)
print("coeeficient = " , model.coef_)
```

```python
# we notice that the accurancy using the Polynomial regression is bigger than the accurancy using the multi-linear and linear regression
# Polynomial regression > multi-linear regression > linear regression 
# as the accurancy is 0.8399 (83.99%) we can say that our model is quiet good
# our date isn't distributed in a form near of a form of a line it's more like a curve this is why in this case the polynomial regression is the best model
```
